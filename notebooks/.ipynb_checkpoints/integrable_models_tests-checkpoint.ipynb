Define the model as sequence of a symplectic transformation T and a normalizing flow for the actions F

```python
T = Chain([OscillatorFlow()])
# T = Chain([SqueezeAndShift(shift_model=MLP()) if i % 2 == 0 
#            else SymplecticExchange() 
#            for i in range(settings['num_bijectors'])])

# Use real NVP, split along the d dimension
# dim_to_split = 1; na = 1; nb = settings['d'] - na
# F = Chain([AffineCoupling(MLP(activation=tf.nn.relu, mode="shift_and_scale"), dim_to_split, [na, nb]),
#            Permute(dim_to_split, [na, nb]),
#            AffineCoupling(MLP(activation=tf.nn.relu, mode="shift_and_scale"), dim_to_split, [na, nb]),
#            Permute(dim_to_split, [na, nb])])
# F = Chain([AdditiveCoupling(MLP(activation=tf.nn.relu, mode="shift"), dim_to_split, [na, nb]),
#            Permute(dim_to_split, [na, nb]),
#            AdditiveCoupling(MLP(activation=tf.nn.relu, mode="shift"), dim_to_split, [na, nb]),
#            Permute(dim_to_split, [na, nb])])

F = Chain([IdFlow()])

# Recall: Chain(A,B) applies first A then B:
model = Chain([AngleInvariantFlow(F), T])
```

```python
def compute_loss(model, hamiltonian, z):
    return tf.reduce_mean( -model.log_jacobian_det(z) + hamiltonian(model(z)) )

def compute_gradients(model, hamiltonian, z):
    with tf.GradientTape() as tape:
        loss = compute_loss(model, hamiltonian, z)
    return tape.gradient(loss, model.trainable_variables), loss

def apply_gradients(optimizer, gradients, variables, global_step=None):
    optimizer.apply_gradients(zip(gradients, variables), global_step=global_step)
```

```python
import time
start_time = time.time()

optimizer = tf.train.AdamOptimizer(1e-4)

global_step = []
losses = []
gradients_norms = []
for epoch in range(settings['train_iters']):
    # Sample
    z = base_dist.sample(settings['batch_size'])
    
    # Compute gradients, loss and update trainable_variables
    gradients, loss = compute_gradients(model, settings['hamiltonian'], z)
    apply_gradients(optimizer, gradients, model.trainable_variables)
    
    # Visualization
    if epoch % 100 == 0:
        global_step.append(epoch)
        losses.append(loss.numpy())
    if epoch % 100 == 0:
        print(epoch, loss.numpy())
        gradients_norms.append([tf.norm(gradients[i]).numpy() for i in range(len(gradients))])        

end_time = time.time()
print("Time elapsed: ", end_time - start_time)
```

```python
start = 0
plt.plot(losses[start:])
```

```python
gradients_norms = np.array(gradients_norms)
rows = gradients_norms.shape[1]//4
cols = 4
f, arr = plt.subplots(rows, cols, figsize=(60,50))
for i in range(rows):
    for j in range(cols):
        c = cols * i + j
        arr[i,j].plot(gradients_norms[:,c])
        arr[i,j].set_title(str(c), fontsize=30)
```

```python
# Test accuracy
z_samples = base_dist.sample(settings['batch_size'])
test_error = compute_loss(model, settings['hamiltonian'], z_samples)
print("Test error: ", test_error)
```
